## Parsing PDF/Scanned PDF

```python
result = await parser.parse_pdf(
    "ADRO_390_410.pdf",
    start_page=16,
    end_page=20,
    parsing_mode=ParsingMode.AGENTIC,
    metadata={
        "company_name": "PT Alamtri Resources",
        "TICKER": "ADRO",
        "year": "2024",
        "document_type": "balance sheet/annual report",
    },
    save_path="output",
)
```

## Structure Output

```python
from typing import List, Optional

from pydantic import BaseModel

system_prompt = """
You are a receipt parser. Your task is to parse the receipt and return the total amount.
"""


class Item(BaseModel):
    name: str
    price: float
    quantity: int


class Receipt(BaseModel):
    items: List[Item]
    total: float
    tax: Optional[float]


parser = DocumentParser(
    vlm_base_url="https://fahmiaziz98--qwen3vl-vllm-server-serve.modal.run/v1",
    vlm_model="Qwen/Qwen3-VL-8B-Instruct",
    vlm_api_key="EMPTY",
    temperature=1.0,
    system_prompt=system_prompt,
    enable_rotate=False,
    rotate_aggressive_mode=True,
    enable_deskew=True,
    debug_dir="debug_image",
)
```

```python
result = await parser.parse_image(
    "complex_invoice.png",
    parsing_mode=ParsingMode.VLM,
    output_schema=Receipt,
    metadata={"document_type": "receipt"},
    save_path="output",
)
```
